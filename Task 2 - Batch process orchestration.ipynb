```python
# -*- coding: utf-8 -*-

# Load data in DB
import os
import pandas as pd
from pymongo import MongoClient
mongodbClient =  MongoClient("mongodb://localhost:27017/")
```

```python
db = mongodbClient["jtc"]

col_dimProducts = db["DimProducts"]
col_dimDates = db["DimDates"]
col_dimStores = db["DimStores"]

dest_path = r'C:\Users\chris\OneDrive\Desktop\jtc\Dumps'

table= pd.read_csv(os.path.join(dest_path, "DimDate.txt"),sep=',', encoding= 'unicode_escape')
data_dict = table.to_dict("records") 
col_dimDates.insert_many(data_dict)

table= pd.read_csv(os.path.join(dest_path, "DimProduct.txt"),sep='\t', encoding= 'unicode_escape')
data_dict = table.to_dict("records") 
col_dimProducts.insert_many(data_dict)


table= pd.read_csv(os.path.join(dest_path, "DimStore.txt"),sep=',', encoding= 'unicode_escape')
data_dict = table.to_dict("records") 
col_dimStores.insert_many(data_dict)
```

```python
# Data flow orchestration to process raw data to generate pipe-delimited load files in the path (/dataextract/yy/MM/dd/factstore).

rootdir = r'C:\Users\chris\OneDrive\Desktop\jtc\data\2009'
dest = r'C:\Users\chris\OneDrive\Desktop\jtc\data'

for subdir, dirs, files in os.walk(rootdir):
    master_df = []
    for file in files:
        print(os.path.join(subdir, file))
        table= pd.read_csv(os.path.join(subdir, file),sep='\t')
        table['DateID'] =  table['Date'].apply(lambda x : x+" 00:00:00")
        table = table[['DateID','StoreID','ProductID','OnHandQuantity','OnOrderQuantity','DaysInStock','MinDayInStock','MaxDayInStock']]
        master_df.append(table)
        directory = os.path.join(dest,"dataextract",subdir.split('\\')[-3],subdir.split('\\')[-2],subdir.split('\\')[-1])

        if not os.path.exists(directory):
            os.makedirs(directory)
            
    if len(master_df)>0:
        factstore = pd.concat(master_df)
        factstore.to_csv(directory+'\\factstore.txt',sep='|',index=False)
```
